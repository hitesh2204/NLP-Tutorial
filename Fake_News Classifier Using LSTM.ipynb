```python
## Importing imporatant libaies.
import pandas as pd
news=pd.read_csv("E://Live Project Dataset//fake_news_train//train.csv")
news.head()
```

```python
### Dropping all null values.
news=news.dropna()
```

```python
### Getting Independent Features.
X=news.drop('label',axis=1)
```

```python
## Getting Dependent Featuure.
y=news['label']
```

```python
### Imprting imporatant libaries.
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## Vocabulary size=
voc_size=10000
```

### One Hot Repesentation.

```python
## copying data into other variable.
messages=news.copy()
```

```python
### Reset the index.
messages.reset_index(inplace=True)
```

```python
### Importing important libries.
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

### Preprocessing Dataset.

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

ps=PorterStemmer()
wordnet=WordNetLemmatizer()

corpus=[]
for i in range(0,len(messages)):
    print(i)
    ## ermoving all unwanted data fom message.
    text=re.sub('[^A-Za-z]',' ',messages['title'][i])
    ## conveting text into lower data.
    text=text.lower()
    ## Splitting the data.
    text=text.split()
    
    ## Finding the all stop words form text.
    text=[wordnet.lemmatize(word) for word in text if not word in stopwords.words('english')]
    text=' '.join(text)
    corpus.append(text)
    
               
                   
```

```python
corpus
```

```python
onehot_rep=[one_hot(words,voc_size)for words in corpus]
print(onehot_rep)
```

### Word Embedding Representation

```python
sent_length=20
embedded_doc=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
len(embedded_doc)
```

```python
## Creating Model
embedded_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedded_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
### Creating final Independent and Dependent variable.
import numpy as np
X_final=np.array(embedded_doc)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
### SPlitting data into training and testing variable.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.33,random_state=0)
```

### Model Training.

```python
### Final Training.
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

### Performance Metrics and Accuracy.

```python
y_pread=model.predict_classes(x_test)
print(y_pread)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_pread,y_test)
print(cm)
print("")
result=accuracy_score(y_pread,y_test)
print(result)
```
