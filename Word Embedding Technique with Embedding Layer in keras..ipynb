```python
### Impotring important libraies.
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### Sentences
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am good boy',
      'I am good developer',
      'undestand the meaning of word',
      'our videos are good']
```

```python
sent
```

```python
## Vocabulary size=
voc_size=10000
```

### One Hot Representation

```python
onehot_rep=[one_hot(words,voc_size)for words in sent]
print(onehot_rep)
```

### Word Embedding Representation

```python
### Impotring important libraries.
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
embedded_doc=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
## Creating Dimention
dim=10
```

```python
## creating the Sequential model.
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
