```python
## Impotring important libaries.
import nltk
import re
import pandas as pd
```

```python
message=pd.read_csv("E://Live Project Dataset//smsspamcollection//SMSSpamCollection",sep='\t',names=['labels','message'])
print(message.head())
```

```python
## importing impoatant NLP libaries.
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
## creating the object of porter stemmer.
ps=PorterStemmer()
wordnet=WordNetLemmatizer()
corpus=[]
for i in range(0,len(message)):
    ## ermoving all unwanted data fom message.
    text=re.sub('[^A-Za-z]','',message['message'][i])
    ## conveting text into lower data.
    text=text.lower()
    ## Splitting the data.
    text=text.split()
    
    ## Finding the all stop words form text.
    text=[wordnet.lemmatize(word) for word in text if not word in stopwords.words('english')]
    text=' '.join(text)
    corpus.append(text)
```

```python
print(corpus)
```

```python
### Creating the Bag of word model.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
X=cv.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(message['labels'],drop_first=True)
print(y)
```

```python
## Splitting data into training and testing featues.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
### training model by using naivy bayes.
from sklearn.naive_bayes import MultinomialNB
spam_detection=MultinomialNB()
spam_detection.fit(x_train,y_train)
```

```python
## perdiction.
y_pread=spam_detection.predict(x_test)
print(y_pread)
```

```python
### Finding accurac and confusion matrix.
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_pread,y_test)
print(cm)
print(" ")
accuracy=accuracy_score(y_pread,y_test)
print(accuracy)
```
